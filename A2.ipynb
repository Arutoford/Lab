```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

from datasetHelper import splitDataset, showMetrics, encodeDataset


data = pd.read_csv ('Desktop/0.csv')

encoded_df = encodeDataset(data)

X_train, X_test, y_train, y_test = splitDataset(encoded_df, targetVariable='26.0')

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

showMetrics(encoded_df, y_test, y_pred)

param_grid = {
    'k_neighbors': [3, 5, 10],
    'weights': ['uniform', 'distance']
}
```
